```python
# importing the library used for data cleaning
import pandas as pd
```

```python
# loading the raw csv file
file = 'file.csv'
usa = pd.read_csv(file, encoding = 'utf-8')
```

```python
usa
```

```python
# starting the cleaning of the csv raw file
# try to make the fields as logical and readable as possible by replacing the unusual characters in the file
usa['2016 land area'] = usa['2016 land area'].str.replace('sq', ' ')
```

```python
usa
```

```python
usa['2016 population density'] = usa['2016 population density'].str.replace('/sq', ' ')
usa['2016 population density'] = usa['2016 population density'].str.replace('mi', ' ')
```

```python
usa
```

```python
usa['Unnamed: 6'] = usa['Unnamed: 6'].str.replace('km2', ' ')
usa['Unnamed: 8'] = usa['Unnamed: 8'].str.replace('/km2', ' ')
```

```python
usa
```

```python
usa['Change'] = usa['Change'].str.replace('%', ' ')
```

```python
usa
```

```python
usa['City'] = usa['City'].str.replace(r"\[.*\]"," ")
```

```python
usa.head(150)
```

```python
usa['2016 land area'] = usa['2016 land area'].str.replace('mi', ' ')
usa['Location'] = usa['Location'].str.replace('°N ', ' ')
```

```python
usa
```

```python
usa['Location'] = usa['Location'].apply(lambda x: x.replace('°W',' '))
```

```python
usa
```

```python
# dividing the location of cities into longitude and respectively latitude
location = pd.DataFrame(usa.Location.str.split(' ',1).tolist(),
                                 columns = ['Latitude','Longitude'])
```

```python
location
```

```python
# concatenation of longitude and latitude columns
usa_city = pd.concat([usa, location], axis = 1)
```

```python
usa_city
```

```python
# Droping the Location column
usa_city.drop(columns='Location', inplace=True)
```

```python
usa_city
```

```python
# renaming of all columns
usa_city.rename(columns = {'Change' : 'Change (%)', '2016 land area' : '2016 land area (mi sq)', 'Unnamed: 6' : '2016 land area (km2)', 
                          '2016 population density' : '2016 population density (hab/mi sq)', 'Unnamed: 8' : '2016 population density (hab/km2)'}, inplace = True)
```

```python
usa_city
```

```python
# Displaying data types and converting them
usa_city.dtypes
```

```python
usa_city['2019 estimate'] = usa_city['2019 estimate'].str.replace(',','')
```

```python
usa_city['2019 estimate'] = usa_city['2019 estimate'].astype('int64')
```

```python
usa_city['2010 Census'] = usa_city['2010 Census'].str.replace(',','')
```

```python
usa_city['2010 Census'] = usa_city['2010 Census'].astype('int64')
```

```python
usa_city['2016 land area (km2)'] = usa_city['2016 land area (km2)'].str.replace(',','')
```

```python
usa_city['2016 population density (hab/mi sq)'] = usa_city['2016 population density (hab/mi sq)'].str.replace(',','')
```

```python
usa_city['2016 population density (hab/km2)'] = usa_city['2016 population density (hab/km2)'].str.replace(',','')
```

```python
usa_city['2016 land area (mi sq)'] = usa_city['2016 land area (km2)'].astype('float64')
```

```python
usa_city['2016 land area (km2)'] = usa_city['2016 land area (km2)'].astype('float64')
```

```python
usa_city['2016 population density (hab/mi sq)'] = usa_city['2016 population density (hab/mi sq)'].astype('int64')
```

```python
usa_city['2016 population density (hab/km2)'] = usa_city['2016 population density (hab/km2)'].astype('int64')
```

```python
usa_city['Longitude'] = usa_city['Longitude'].astype('float64')
```

```python
usa_city['Latitude'] = '-' + usa_city['Latitude'].astype(str)
```

```python
usa_city['Latitude'] = usa_city['Latitude'].astype('float64')
```

```python
usa_city
```

```python
# droping the Change % column
usa_city.drop(columns = 'Change (%)', inplace = True)
```

```python
usa_city
```

```python
# adding and calculation of Population_difference column
usa_city['Population_difference'] = usa_city.apply(lambda x : x['2019 estimate'] - x['2010 Census'], axis = 1)
```

```python
usa_city
```

```python
# Saving the new cleaned and processed file
usa_city.to_csv('united_states_cities.csv',index = False)
```
